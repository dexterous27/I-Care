```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset_survey.csv')
df1 = df.copy()
df1
```

```python
# Drop column  from the DataFrame

columns_to_drop = ['Timestamp','Program Name','Term you are currently enrolled in.']
df1 = df1.drop(columns_to_drop, axis=1)
# Display the modified DataFrame
print("DataFrame after dropping columns:")
print(df1)
```

```python
df1 = pd.DataFrame(df1)
df1
```

```python
# Reset the index to convert it into a column
df1.reset_index(inplace=True)

# Use the melt function to convert columns into rows
melted_df = pd.melt(df1, id_vars='index', var_name='Questions', value_name='Answers')

# Drop the 'index' column
melted_df.drop('index', axis=1, inplace=True)

# Display the new DataFrame
print(melted_df)
```

```python
df2 = pd.DataFrame(melted_df)
df2.head(20)
```

```python
column_names = df2.columns
print(column_names)

# Check column names using the .head() method
column_names_head = df2.head(0)
print(column_names_head)


```

```python
df2['Row'] = df2.groupby('Questions').cumcount() + 1

# Pivot the DataFrame to create different columns for each row under the same question
pivot_df = df2.pivot(index='Questions', columns='Row', values='Answers')

# Reset the index and rename the columns
pivot_df.reset_index(inplace=True)
pivot_df.columns = ['Questions'] + [f'Answer_{col}' for col in pivot_df.columns[1:]]

# Display the pivoted DataFrame
print(pivot_df)
```

```python
df2 = pd.DataFrame(pivot_df)
```

```python


# Select a subset of columns to visualize
columns_to_visualize = column_names_head # Add your column names here

# Visualize numerical columns
numerical_columns = df2.select_dtypes(include=['int64', 'float64']).columns
numerical_columns = [col for col in numerical_columns if col in columns_to_visualize]
for column in numerical_columns:
    plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
    sns.histplot(data=df2, x=column)
    plt.title(f"Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()

# Visualize categorical columns
categorical_columns = df2.select_dtypes(include=['object']).columns
categorical_columns = [col for col in categorical_columns if col in columns_to_visualize]
for column in categorical_columns:
    plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
    sns.countplot(data=df2, x=column)
    plt.title(f"Count of {column}")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.show()
```

```python
plt.figure(figsize=(15, 8))

# Plot each column against the "Questions" column
for column in df2.columns:
    if column != "Questions":  # Exclude the "Questions" column from plotting
        plt.plot(df2["Questions"], df2[column], label=column)

# Set the x-axis and y-axis labels
plt.xlabel("Questions")
plt.ylabel("Column Values")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Apply vectorization to the "Questions" column
vectorized_questions = vectorizer.fit_transform(df2["Questions"])

# Convert the vectorized data into a DataFrame
df_vectorized = pd.DataFrame(vectorized_questions.toarray(), columns=vectorizer.get_feature_names())

# Add the "Questions" column back to the DataFrame
df_vectorized["Questions"] = df2["Questions"]

# Plot the vectorized data
plt.figure(figsize=(15, 8))

# Plot each column (except "Questions") against the "Questions" column
for column in df_vectorized.columns:
    if column != "Questions":
        plt.plot(df_vectorized["Questions"], df_vectorized[column], label=column)

# Set the x-axis and y-axis labels
plt.xlabel("Questions")
plt.ylabel("Vectorized Values")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Add a legend
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Show the plot
plt.show()
```
